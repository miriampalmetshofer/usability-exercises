```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Read data

```python
df_pc_mouse = pd.read_csv('./data/cleaned/pc-mouse.csv', sep=';')
df_smartphone = pd.read_csv('./data/cleaned/smartphone.csv', sep=';')
df_tablet_stylus = pd.read_csv('./data/cleaned/tablet-stylus.csv', sep=';')
df_tablet_touch = pd.read_csv('./data/cleaned/tablet-touch.csv', sep=';')

display(df_pc_mouse)
display(df_smartphone)
display(df_tablet_stylus)
display(df_tablet_touch)
```

```python
df_smartphone = df_smartphone.drop(columns=['CorrectedTextErrorRate(%)', 'Entry_mode'])
# rename column TranscribedTextErrorRate(%) to ErrorRate(%)
df_smartphone = df_smartphone.rename(columns={'TranscribedTextErrorRate(%)': 'Error_rate_(%)'}) 

df_smartphone['type'] = 'smartphone'
df_tablet_touch['type'] = 'tablet-touch'
df_smartphone_tablet_touch = pd.concat([df_smartphone, df_tablet_touch], ignore_index=True)
df_smartphone_tablet_touch
```

```python
# combine the tablet dfs, reset the index, but add a column defining the type of tablet
df_tablet_stylus['type'] = 'stylus'
df_pc_mouse['type'] = 'mouse'
df_tablet = pd.concat([df_tablet_stylus, df_pc_mouse], ignore_index=True)
# remove nr 30
df_tablet = df_tablet.drop(30)

display(df_tablet)
```

```python
df_tablet['Session_num'] = df_tablet.groupby('Participant').cumcount() + 1
df_smartphone_tablet_touch['Session_num'] = df_smartphone_tablet_touch.groupby('Participant').cumcount() + 1

```

```python
def plot_metric_boxplot(df, metric, palette='husl', df_name='SmartphoneTabletTouch'):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Session_num', y=metric, hue='type', palette=palette)

    plt.xlabel('Session Number')
    plt.ylabel(metric)
    plt.title(f'{metric} Development Over Sessions by Input Type')
    plt.legend(title='Input Type')

    plt.savefig(f'figures/{df_name}_{metric}_boxplot.png', bbox_inches='tight', dpi=150)

    plt.show()

print('PC Mouse and Tablet Stylus')
plot_metric_boxplot(df_tablet, 'Speed_(wpm)', 'tab10', 'PcMouseTabletStylus')
plot_metric_boxplot(df_tablet, 'Time_(s)', 'tab10', 'PcMouseTabletStylus')
plot_metric_boxplot(df_tablet, 'Error_rate_(%)', 'tab10', 'PcMouseTabletStylus')
plot_metric_boxplot(df_tablet, 'KSPC', 'tab10', 'PcMouseTabletStylus')

plot_metric_boxplot(df_tablet, 'Speed_(wpm)')
plot_metric_boxplot(df_smartphone_tablet_touch, 'Speed_(wpm)')
plot_metric_boxplot(df_smartphone_tablet_touch, 'Time_(s)')
plot_metric_boxplot(df_smartphone_tablet_touch, 'Error_rate_(%)')
plot_metric_boxplot(df_smartphone_tablet_touch, 'KSPC')
```

```python


def plot_metric_lineplot(df, metric, palette='husl', df_name='SmartphoneTabletTouch'):
    df_avg = df.groupby(['Session_num', 'type'])[metric].median().reset_index()

    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df_avg, x='Session_num', y=metric, hue='type', marker='o', palette=palette)
    plt.xticks(range(1, df_avg['Session_num'].max() + 1))

    plt.xlabel('Session Number')
    plt.ylabel(metric)
    plt.title(f'{metric} Development Over Sessions by Input Type')
    plt.legend(title='Input Type')

    plt.savefig(f'figures/{df_name}_{metric}.png', bbox_inches='tight', dpi=150)

    plt.show()

print('PC Mouse and Tablet Stylus')
plot_metric_lineplot(df_tablet, 'Time_(s)', 'tab10', 'PcMouseTabletStylus')
plot_metric_lineplot(df_tablet, 'Speed_(wpm)', 'tab10', 'PcMouseTabletStylus')
plot_metric_lineplot(df_tablet, 'Error_rate_(%)', 'tab10', 'PcMouseTabletStylus')
plot_metric_lineplot(df_tablet, 'KSPC', 'tab10', 'PcMouseTabletStylus')

print('Smartphone and Tablet Touch')
plot_metric_lineplot(df_smartphone_tablet_touch, 'Error_rate_(%)')
plot_metric_lineplot(df_smartphone_tablet_touch, 'Speed_(wpm)')
plot_metric_lineplot(df_smartphone_tablet_touch, 'Time_(s)')
plot_metric_lineplot(df_smartphone_tablet_touch, 'KSPC')
```
